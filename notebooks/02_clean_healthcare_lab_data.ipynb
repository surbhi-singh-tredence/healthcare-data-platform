```python
from src.transformations.silver_cleaning import clean_data, write_silver_table
from pyspark.sql.functions import col
```

```python
BRONZE_TABLE = "dbdemos.bronze.healthcare_qa_raw"
SILVER_TABLE = "dbdemos.silver.healthcare_qa_clean"


df = spark.table(BRONZE_TABLE)

clean_df = clean_data(df)

def enforce_schema(df):

    df = df.select(
        col("label_id").cast("string"),
        col("sample_number").cast("int"),
        col("lot_name").cast("string"),
        col("lot_number").cast("int"),
        col("item_description").cast("string"),
        col("date_completed").cast("timestamp"),
        col("test_number").cast("int"),
        col("name").cast("string"),
        col("product").cast("string"),
        col("status").cast("string"),
        col("entry").cast("string"),
        col("product_grade").cast("string"),
        col("units").cast("string"),
        col("stage").cast("string"),
        col("x_method").cast("string"),
        col("reported_name").cast("string"),
        col("sampled_date").cast("timestamp"),
        col("description").cast("string"),
        col("reportable").cast("string"),
        col("x_disposed").cast("string"),
        col("result_number").cast("int")
    )
    return df

clean_df = enforce_schema(clean_df)
write_silver_table(clean_df, SILVER_TABLE)
```
